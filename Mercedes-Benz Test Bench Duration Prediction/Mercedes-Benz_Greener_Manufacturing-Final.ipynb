```python
# Perform dimensionality reduction.
# Predict your test_df values using XGBoost.

from sklearn.decomposition import PCA
```

```python
X.shape, test_data.shape
```

```python
n_comp = 12
pca = PCA(n_components=n_comp, random_state=111)
pca2_results_train = pca.fit_transform(X)
pca2_results_test = pca.transform(test_data)
```

```python
pca2_results_train.shape, pca2_results_test.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_sample, y_train, y_sample = train_test_split(pca2_results_train, y, test_size=0.2, random_state=123)
```

```python
X_train.shape, X_sample.shape, y_train.shape, y_sample.shape
```

```python
from sklearn.linear_model import LinearRegression
import xgboost as xgb
from sklearn.metrics import r2_score 
```

```python
# we will apply Xgboost regression and see how it varies the results
```

```python
d_train = xgb.DMatrix(data=X_train,label=y_train)
d_sample = xgb.DMatrix(data=X_sample, label=y_sample)

d_test = xgb.DMatrix(pca2_results_test)
```

```python
params = {}
params['objective'] = 'reg:linear'
params['eta'] = 0.02
params['max_depth'] = 4

def xgb_r2_score(preds, dtrain):
    labels = dtrain.get_label()
    return 'r2', r2_score(labels, preds)
```

```python
watchlist = [(d_train, 'train'), (d_sample, 'sample')]

clf = xgb.train(params, d_train, 
                1000, watchlist, early_stopping_rounds=50, 
                feval=xgb_r2_score, maximize=True, verbose_eval=10)
```

```python
p_test = clf.predict(d_test)

sub = pd.DataFrame()
sub['ID'] = test['ID']
sub['y'] = p_test
sub.to_csv('xgb.csv', index=False)

sub.head()
```
