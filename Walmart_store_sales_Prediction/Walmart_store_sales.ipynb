```python
# 1. Which store has maximum sales?
# Assumption 1: Individually Weekly_sales by Store in 2010, 2011, 2012
# Assumption 2: Combining Sales of each stores for individual year 2010, 2011 & 2012 respectively. 
# Assumption 3: Combining Sales of each stores for all the 3 years.
```

```python
store_sales_2010 = df_Walmart[df_Walmart.Year==2010]['Weekly_Sales'].groupby(df_Walmart['Store']).max()
store_sales_2011 = df_Walmart[df_Walmart.Year==2011]['Weekly_Sales'].groupby(df_Walmart['Store']).max()
store_sales_2012 = df_Walmart[df_Walmart.Year==2012]['Weekly_Sales'].groupby(df_Walmart['Store']).max()
```

```python
store_sales_2010
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(store_sales_2010.index, store_sales_2010.values)
sns.lineplot(store_sales_2011.index, store_sales_2011.values)
sns.lineplot(store_sales_2012.index, store_sales_2012.values)
plt.grid()
plt.legend(['2010','2011','2012'], loc='best', fontsize=14)
plt.title('Average_Weekly_Sales_of_Store - Year', fontsize=16)
plt.xlabel('Store', fontsize=20)
plt.ylabel('Individual_Weekly_Sales', fontsize=20)
plt.xticks(np.arange(1,46, step=1))
plt.show()
```
