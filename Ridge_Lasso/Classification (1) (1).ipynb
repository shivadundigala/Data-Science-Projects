```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
print(cancer.keys())
```

```python
print(cancer.DESCR)
```

```python
import pandas as pd
data=pd.DataFrame(cancer.data,columns=cancer.feature_names)
data['Target']=cancer.target
```

```python
data.head()
```

```python
data.Target.value_counts()
```

```python
from sklearn.model_selection import train_test_split as split
train,test=split(data,test_size=.30,random_state=2020)
X=train.drop(columns="Target")
y=train.Target
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(X,y)
```

```python
pred=lg.predict(test.drop(columns="Target"))
```

```python
pred
```

```python
pred_p=lg.predict_proba(test.drop(columns="Target"))
```

```python
pred_p
```

```python
from sklearn.metrics import classification_report, roc_curve, roc_auc_score
```

```python
pd.crosstab(columns=test.Target,index=pred)
```

```python
fpr,tpr,thres=roc_curve(y_true=test.Target,y_score=pred_p[:,1],pos_label=1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.plot(fpr,tpr)
plt.plot([0,1],[0,1])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.show()
```

```python
roc_auc_score(y_true=test.Target,y_score=pred_p[:,1])
```
